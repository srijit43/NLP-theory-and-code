# Supplemental Data Cleaning: Using Stemming

### Test out Porter stemmer

```python
import nltk

ps = nltk.PorterStemmer()
```

```python
print(ps.stem('grows'))
print(ps.stem('growed'))
print(ps.stem('growing'))  #stem is the method to reduce it to grow
```

```python
print(ps.stem('run'))
print(ps.stem('running'))
print(ps.stem('runner'))
```

Import other dependencies

```python
import pandas as pd
import re
import string
pd.set_option('display.max_colwidth',100)

nltk.download('stopwords')
```

```python
data = pd.read_csv('SMSSpamCollection.tsv',sep='\t')
data.columns = ['label','body_text']
data.head()
```

Data cleaning function collapsed into one function

```python
stopwords = nltk.corpus.stopwords.words('english')

def clean_data(text):
  text = "".join([word for word in text if word not in string.punctuation ]) #remove punctuation
  tokens = re.split('\W+',text)  #tokenize
  text = [word for word in tokens if word not in stopwords] #remove stopwords
  return text
```

```python
data['body_text_nonstop'] = data['body_text'].apply(lambda x:clean_data(x.lower()))
```

```python
data.head()
```

Now let us do the stemming

```python
def stemming(tokenized_text):
  text = [ps.stem(word) for word in tokenized_text]
  return text

data['body_text_stemmed'] = data['body_text_nonstop'].apply(lambda x:stemming(x))

data.head()


```

### Read in raw text

```python
import pandas as pd
import re
import string
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']

data.head()
```

### Clean up text

```python
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text

data['body_text_nostop'] = data['body_text'].apply(lambda x: clean_text(x.lower()))
```
