# Feature Engineering: Feature Creation

### Read in text

```python
import pandas as pd

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']
```

### Create feature for text message length

```python
data['body_length'] = data['body_text'].apply(lambda x:len(x) - x.count(" ")) #only gives total length of characters and not white spaces
data.head()
```

### Create feature for % of text that is punctuation

This is because real text messages use less punctuations than spam mails(hypothesis)

```python
import string

def punctuationPercent(text):
  count = sum([1 for char in text if char in string.punctuation])   #returns 1 if char is a punctuation and we do sum of all punctuations in a text( loops for the whole corpus)
  return round(count/(len(text) - text.count(" ")),3)*100       #returns % rounded to 3 decimals
```

Apply this function as lambda function to body_text

```python
data['punctuation_percent_in_text'] = data['body_text'].apply(lambda x:punctuationPercent(x))
data.head()
```
