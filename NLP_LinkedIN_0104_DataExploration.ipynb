Today we are going to do some data exploration

Let us begin by importing dependencies

```python
import pandas as pd
import numpy as np
```

```python
fullCorpus = pd.read_csv("SMSSpamCollection.tsv",sep = '\t', header = None)
```

```python
fullCorpus.head()
```

```python
fullCorpus.columns = ['Labels','Body_Text']
```

```python
fullCorpus.head()
```

What is the shape of the data?

```python
print("Input data has {} rows and {} columns".format(len(fullCorpus),len(fullCorpus.columns)))
```

How many spam and ham ?

```python
#the split is important to note as it will lead to a different ML appraoch
```

```python
print("Out of {} rows,There are {} spam and {} ham".format(len(fullCorpus),
                                            len(fullCorpus[fullCorpus['Labels']=='spam']),
                                            len(fullCorpus[fullCorpus['Labels']=='ham'])))
```

How much missing data is there?

```python
print("Number of nulls in the label:{}".format(fullCorpus['Labels'].isnull().sum()))
print("Number of nulls in the body:{}".format(fullCorpus['Body_Text'].isnull().sum()))
```
