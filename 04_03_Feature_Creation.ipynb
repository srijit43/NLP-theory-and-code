# Feature Engineering: Feature Creation

### Read in text

```python
import pandas as pd

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']
```

### Create feature for text message length

```python
data['body_len'] = data['body_text'].apply(lambda x: len(x) - x.count(" "))

data.head()
```

### Create feature for % of text that is punctuation

```python
import string

def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")), 3)*100

data['punct%'] = data['body_text'].apply(lambda x: count_punct(x))

data.head()
```

### Evaluate created features

```python
from matplotlib import pyplot
import numpy as np
%matplotlib inline
```

###We are going to see the 1) body length distribution and 2) punctuation percentage for spam and not spam

```python
bins = np.linspace(0,200,40)  #start, end , cutpoint

pyplot.hist(data[data['label']=='spam']['body_len'],bins,alpha=0.5,density = True, label = 'spam')  #we use normed so that ham does not look like huge beside spam (considering more ham than spam)
pyplot.hist(data[data['label']=='ham']['body_len'],bins,alpha=0.5,density = True, label = 'ham')
pyplot.legend(loc = 'upper left')
pyplot.show()
```

```python
bins2 = np.linspace(0,50,40)  #start, end , cutpoint

pyplot.hist(data[data['label']=='spam']['punct%'],bins2,alpha=0.5,density = True, label = 'spam')  #we use normed so that ham does not look like huge beside spam (considering more ham than spam)
pyplot.hist(data[data['label']=='ham']['punct%'],bins2,alpha=0.5,density = True, label = 'ham')
pyplot.legend(loc = 'upper right')
pyplot.show()
```
