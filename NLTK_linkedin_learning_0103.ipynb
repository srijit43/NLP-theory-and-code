```python
raw_data = open('SMSSpamCollection.tsv').read()

#print raw data
raw_data[0:500]
```

Job 1 : In this case replace \n with \t this converts it into readable text for python: ✔

```python
parsedData = raw_data.replace('\t','\n').split('\n')  #returns a list
```

```python
print(parsedData[0:5])
```

Viola looks good right?? 🦁

```python
labelList = parsedData[0::2]  #every other place starting from 0 to get the speaker
textList = parsedData[1::2]  #next alternate to pick the text
```

```python
print(labelList[0:5])
print(textList[0:5])
```

```python
import pandas as pd
```

```python
#We are going to create a data frame
```

```python
fullCorpus = pd.DataFrame({
    'labels':labelList,
    'body_list':textList
})
```

Let us see the issue

```python
print(len(labelList))
print(len(textList))
```

```python
#to troubleshoot we print the last 5 elements of labelList
print(labelList[-5:])
```

```python
fullCorpus = pd.DataFrame({
    'labels':labelList[:-1],
    'body_list':textList
})                                       #grab all but the last one
```

Now we have fullCorpus which has key as speaker and values as text
