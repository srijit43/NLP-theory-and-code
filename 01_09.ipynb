# NLP Basics: Implementing a pipeline to clean text

### Pre-processing text data

Cleaning up the text data is necessary to highlight attributes that you're going to want your machine learning system to pick up on. Cleaning (or pre-processing) the data typically consists of a number of steps:
1. **Remove punctuation**
2. **Tokenization**
3. **Remove stopwords**
4. Lemmatize/Stem

The first three steps are covered in this chapter as they're implemented in pretty much any text cleaning pipeline. Lemmatizing and stemming are covered in the next chapter as they're helpful but not critical.

```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t', header=None)
data.columns = ['label', 'body_text']

data.head()
```

1. Remove punctuation

```python
import string
```

```python
string.punctuation
```

```python
def remove_punct(text):
  text_nopunct = "".join([char for char in text if char not in string.punctuation])  #return char for each char in text if not in string.punctuation , join actually joins the individual chars
  return text_nopunct

data['body_text_clean'] = data['body_text'].apply(lambda x:remove_punct(x))
```

```python
data.head()
```

2. TOKENIZATION --> get word tokens

```python
import re
```

```python
def tokenize(text):
  tokens = re.split('\W+',text)
  return tokens

data['body_text_tokenized'] = data['body_text'].apply(lambda x:tokenize(x.lower()))
data.head()
```

3. Removing stopwords --> this we do to take our data to the pre vectorized form

```python
import nltk
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
  text = [word for word in tokenized_list if word not in stopwords]
  return text

data['body_text_nonstop'] = data['body_text_tokenized'].apply(lambda x:remove_stopwords(x))
data.head()
```
